## Importing Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split #Used to split training and testing data
from sklearn.linear_model import LogisticRegression   # To import logistic regression
from sklearn.metrics import accuracy_score #Measure the accuracy of model
```

# Data Collection and Data Preprocessing

```python
#Loading data into DataFrame
sonar_data = pd.read_csv("/content/Copy of sonar data.csv" , header = None)
```

```python
sonar_data.head()
```

```python
pd.set_option("display.max_columns" , None)
```

```python
sonar_data.head()
```

```python
sonar_data.describe()
```

```python
sonar_data[60].value_counts() #More the data more accurate prediction   # M ---> Mine    R ---> Rock
```

```python
sonar_data.groupby(60).mean()
```

```python
#Seprating data and label
X = sonar_data.drop(columns=60 , axis=1)
Y = sonar_data[60]
```

```python
print(X)
```

```python
print(Y)
```

## Train and Test Data

```python
X_train , X_test , Y_train , Y_test  = train_test_split(X , Y , test_size = 0.1 , stratify=Y , random_state = 1)
```

```python
print("Shape of X : " , X.shape)
print("Shape of training dataset : " , X_train.shape)
print("Shape of test dataset : " , X_test.shape)
```

```python
print(X_train)
print(Y_train)
```

# Model training using Logistic regression

```python
model = LogisticRegression()
```

```python
#Training the Logistic Regression model with the training data
model.fit(X_train , Y_train)
```

# Model Evaluation

```python
#accuracy of training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction , Y_train)
```

```python
print("Accuracy of training data: " , training_data_accuracy)
```

```python
#accuracy of testing data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction , Y_test)
```

```python
print("Accuracy of testing data: " , test_data_accuracy)
```
